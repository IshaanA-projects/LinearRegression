```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import math
```

```python
from google.colab import files

uploaded = files.upload()
file_name = next(iter(uploaded))
df = pd.read_csv(file_name)
```

```python
df.head()
```

```python
df = df[["Open", "High", "Low", "Close", "Volume"]]
df.head()
```

```python
df["R_Range"] = (df['High']-df['Low'])/df['Close'] * 100  # Creates column indicating relative range of the stock price
df["PCT_Change"] = (df['Close']-df['Open'])/df['Close'] * 100 # Creates column indicating percentage change on a day
df["Volatility"] = df['PCT_Change'].rolling(window = 20).std() # Creates column indicating stock price volatility
df = df[["Close", "Volume", "R_Range", "PCT_Change", "Volatility"]] # Keeps only useful columns
df.dropna(inplace = True)
df.head()
```

```python
forecast_column = "Close" # Defines the column we want to predict
forecast_count = math.ceil(len(df)*0.01) # Defines how far ahead we want to predict
df.fillna("-9999", inplace = True) # Fills in blank data with anomalies that have no big effect
forecast_count
```

```python
df["label"] = df['Close'].shift(forecast_count) # Creates the label as the the stock price a few weeks in the future
df.dropna(inplace = True) # Makes sure only columns with a valid label are used
df.tail()
```

```python
X = np.array(df.drop(["label"], axis = 1)) # Creates features
y = np.array(df["label"]) # Creates label
```

```python
X = preprocessing.scale(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2) # Seperates data into training set and testing set
```

```python
clf = LinearRegression() # Creates a linear regression model
clf.fit(X_train, y_train) # Train it on the testing set
accuracy = clf.score(X_test, y_test) # Calculates accuracy
accuracy
```
